### 7. Who has a higher speed stat, Eevee or Pikachu?

```python
for result in data1['results']:
    if result['name'] == "pikachu":
        url6 = result['url']
    elif result['name'] == "eevee":
        url7 = result['url']
url6,url7
```

```python
response6 = requests.get(url6)
response6
```

```python
response7 = requests.get(url7)
response7
```

```python
data6 = response6.json()
data6.keys()
```

```python
data7 = response7.json()
data7.keys()
```

```python
for stat in data6['stats']:
    if stat['stat']['name'] == 'speed':
        pikachu_speed = stat['base_stat']
pikachu_speed
```

```python
for stat in data7['stats']:
    if stat['stat']['name'] == 'speed':
        eevee_speed = stat['base_stat']
eevee_speed
```

```python
if pikachu_speed > eevee_speed:
    print("Pikachu has a higher speed stat.")
elif pikachu_speed < eevee_speed:
    print("Eevee has a higher speed stat.")
else:
    print("Pikachu and Eevee have the same speed stat.")
```
